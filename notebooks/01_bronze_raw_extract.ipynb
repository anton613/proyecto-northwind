```python
import sys
from pathlib import Path
sys.path.append(str(Path('..').resolve()))

from src.database import conexion_mysql, extraer_tabla
from src.config import RAW_PATH, TABLAS_RAW, get_raw_file, verificar_extraccion
import pandas as pd
```

```python
# ==========================================
# CONFIGURACIÓN (ahora viene de src.config)
# ==========================================

print(f"📁 Ruta de datos crudos: {RAW_PATH}")

# ==========================================
# EXTRAER TODAS LAS TABLAS
# ==========================================

print(f"🎯 Iniciando extracción de {len(TABLAS_RAW)} tablas...\n")

with conexion_mysql() as conn:
    for tabla in TABLAS_RAW:
        try:
            datos, columnas = extraer_tabla(conn, tabla)
            
            if datos:
                df = pd.DataFrame(datos, columns=columnas)
                
                # ✅ USAR FUNCIÓN DE CONFIG
                output_file = get_raw_file(tabla)
                df.to_csv(output_file, index=False, encoding='utf-8')
                print(f"💾 {tabla:<30} ({df.shape[0]:>3} filas, {df.shape[1]:>2} columnas)")
            else:
                print(f"⚠️  {tabla:<30} (Tabla vacía)")
            
        except Exception as e:
            print(f"❌ {tabla:<30} Error: {e}")

# ==========================================
# VERIFICACIÓN FINAL
# ==========================================

print(f"\n{'='*50}")
print("📊 RESUMEN DE EXTRACCIÓN")
print(f"{'='*50}")

estado = verificar_extraccion()
print(f"✅ Tablas extraídas: {estado['total_extraidas']}/{estado['total_esperadas']}")

if estado['faltantes']:
    print(f"⚠️  Faltantes: {', '.join(estado['faltantes'])}")
else:
    print("🎉 ¡Todas las tablas extraídas correctamente!")
```
